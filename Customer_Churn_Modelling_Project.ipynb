```python
#Importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importing the Dataset
df=pd.read_csv('Churn_Modelling.csv')
```

```python
df.describe()
```

```python
X = df.drop(labels=['CustomerId', 'Surname', 'RowNumber', 'Exited'], axis = 1)
y = df['Exited']
```

```python
#Convert Categorical Data into Numerical Data

from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
X['Gender'] = label.fit_transform(X['Gender'])
X['Geography'] = label.fit_transform(X['Geography'])
X.head(10)
```

```python
#Dummy Variable Trap
X = pd.get_dummies(X, drop_first=True, columns=['Geography'])
X.head(10)
```

```python
#Splitting Test Set and Train Set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0,)
```

```python
#Feature Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Build ANN
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from keras.layers import Dense 
```

```python
model = Sequential()
model.add(Dense(X.shape[1], activation='relu', input_dim =11))
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
X.shape[1]
```

```python
model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics=['accuracy'])
```

```python
# Model fitting
model.fit(X_train, y_train.to_numpy(), batch_size = 10, epochs = 100)
```
